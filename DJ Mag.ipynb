```python
main_url = 'https://djmag.com/top100djs'

import requests
result = requests.get(main_url)

result.text[:1000]
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(result.text, 'html.parser')

print(soup.prettify()[:1000])
```

```python
def getAndParseURL(url):
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    return(soup)
```

```python
soup.find('article').div.a.get('href')
```

## Get urls for all the top 100 DJs

```python
main_page_products_urls = [x.div.a.get('href') for x in soup.findAll("article")]

print(str(len(main_page_products_urls)) + " fetched products URLs")
print("One example:")
main_page_products_urls[17]
```

```python
main_url_new = 'https://djmag.com'

dj_url = [main_url_new + x.div.a.get('href') for x in soup.findAll('article')]
dj_url = dj_url

print(str(len(dj_url)) + " fetched URLs")
print("Some examples:")
dj_url[:5]
```
